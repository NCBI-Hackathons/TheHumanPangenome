## `02_run_assembler`: A Notebook for running the Peregrine assembler

This notebook is used in the development for the benchmark assembly. A public docker image registry.hub.docker.com/cschin/peregrine:mhc_hg002_20200325 contains the running environment, data, and the script to reproduce the benchmark assembly.
 
This notebook only contains the initial assembly part. The assembly contig generated by this notebook does not resolve the repeats around the C4A region. Please see the docker image for the script and code to reproduce our results resolving the repeats.


```python
%%bash
dx download -f /20200316_asm_for_revision/data/H1.fa
dx download -f /20200316_asm_for_revision/data/H2.fa
dx download -f /20200316_asm_for_revision/data/unphased.fa
mkdir -p  /home/dnanexus/phased_asm_20200320
cd /home/dnanexus/phased_asm_20200320
mkdir -p  /home/dnanexus/phased_asm_20200320/phased_reads/
cp /opt/notebooks/H1.fa /opt/notebooks/H2.fa /opt/notebooks/unphased.fa  /home/dnanexus/phased_asm_20200320/phased_reads/
```

```python
%%bash
cd /home/dnanexus/phased_asm_20200320/
echo /wd/phased_reads/H1.fa > reads.lst
echo /wd/phased_reads/unphased.fa >> reads.lst

docker run --rm -v $PWD:/wd --user $(id -u):$(id -g) \
        --entrypoint /opt/entry_dev.sh cschin/peregrine:0.1.5.2 \
            asm /wd/reads.lst 4 4 4 4 4 4 4 4 4 --shimmer-r 3 \
            --with-consensus \
            --best_n_ovlp 8 \
            --output /wd/asm-HG002-MHC-H1-r3.pg0.1.5.2/
            
cp reads.lst asm-HG002-MHC-H1-r3.pg0.1.5.2/
```

```python
%%bash
cd /home/dnanexus/phased_asm_20200320/
echo /wd/phased_reads/H2.fa > reads.lst
echo /wd/phased_reads/unphased.fa >> reads.lst


docker run --rm -v $PWD:/wd --user $(id -u):$(id -g) \
        --entrypoint /opt/entry_dev.sh cschin/peregrine:0.1.5.2 \
            asm /wd/reads.lst 4 4 4 4 4 4 4 4 4 --shimmer-r 3 \
            --with-consensus \
            --best_n_ovlp 8 \
            --output /wd/asm-HG002-MHC-H2-r3.pg0.1.5.2/
            
cp reads.lst asm-HG002-MHC-H2-r3.pg0.1.5.2/
```
